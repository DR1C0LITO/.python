```python
import matplotlib.pyplot as plt

meses = ['Janeiro', 'Fevereiro', 'Março', 'Abril', 'Maio', 'Junho', 'Julho', 'Agosto', 'Setembro', 'Outubro', 'Novembro', 'Dezembro']
vendas = [1000, 1500, 1200, 1800, 2000, 2200, 2500, 3000, 2800, 3500, 4000, 4500]

plt.plot(meses, vendas,marker='o')
plt.xlabel('Mês')
plt.ylabel('vendas')
plt.title('crescimento')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
cidades = ['rio de janeire', 'salvador', 'Belo horizonte']
temperatura = [42,45,21]
format= plt.bar(cidades, temperatura,color='cyan')
plt.bar_label  (format, labels=temperatura, padding=3, fontsize=10, fontweight='bold')
plt.xlabel('Cidades')
plt.ylabel('temperatura media')
plt.title('comparação nas cidades')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
pizzas=[130,120,533,100,300]

labels =['1,2,3,4,5']
dicionario={'pizzas':pizzas, 'labels':labels}
grafico_pizza = pd.DataFrame(dicionario)
plt.pie(x=grafico_pizza.pizzas,labels = grafico_pizza.labels)
plt.xlabel("pizza")
plt.ylabel("hotdog")
plt.show
```
