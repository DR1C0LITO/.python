```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("pagamentos.csv",sep=";",encoding="utf-8")

def grafico_barra():
    plt.figure()
    plt.bar(df["Nome"],df["Salario Liquido (R$)"])
    plt.xticks(rotation=45)
    plt.title("Salário Líquido do professor")
    plt.xlabel("Professor")
    plt.ylabel('Salário Líquido')
    plt.tight_layout()
    plt.show()
def grafico_pizza():
    plt.figure()
    plt.pie(df["Salario Liquido (R$)"], labels=df["Nome"], autopct="%1.1f%%")
    plt.title("Distribuição do Salário")
    plt.show()

def grafico_histograma():
    plt.figure()
    plt.hist(df["Salario Liquido (R$)"], bins=5)
    plt.title("salario")
    plt.xlabel("faixa salarial")
    plt.ylabel("Quantidade")
    plt.show()
def menu():
    while True:
        print("Escolha um gráfico para visualizar:")
        print("1. Gráfico de Barras")
        print("2. Gráfico de Pizza")
        print("3. Histograma")
        print("4. Sair")
        escolha = input("Digite o número da opção desejada: ")
        if escolha == "1":
            grafico_barra()
        elif escolha == "2":
            grafico_pizza()
        elif escolha == "3":
            grafico_histograma()
        elif escolha == "4":
            print("Saindo do programa.")
            break
        else:
            print("Opção inválida. Tente novamente.")
menu()
```
